```python
import cv2
from matplotlib import pyplot as plt

def grayscale_and_display_in_notebook(image_path):
    # Load the color image
    color_image = cv2.imread(image_path)

    # Convert the color image to grayscale
    grayscale_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    # Display the grayscale image in the notebook
    plt.imshow(grayscale_image, cmap='gray')
    plt.axis('off')  # Optional: Hide axis
    plt.show()

# Example usage
input_image_path = 'C:\\Users\\jagad\\OneDrive\\Pictures\\CT\\hostel.jpg'
grayscale_and_display_in_notebook(input_image_path)
```
